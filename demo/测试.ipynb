```python
import pandas as pd  
  
# 从第一个表格中抽取指定列  
file1_columns = ["年份", "季度", "证券代码", "公司发布财报的日期", "财报统计的季度的最后一天", "流动比率", "速动比率", "现金比率", "总负债同比增长率", "资产负债率", "权益乘数"]  
file1 = pd.read_excel("1.xlsx", usecols=file1_columns)  
file1
```

```python
# 从第二个表格中抽取指定列  
file2_columns = ["流动资产除以总资产", "有形资产除以总资产", "已获利息倍数", "经营活动产生的现金流量净额除以营业收入", "经营性现金净流量除以净利润", "	经营性现金净流量除以营业总收入"]  
file2 = pd.read_excel("2.xlsx", usecols=file2_columns)  
file2
```

```python
# 从第三个表格中抽取指定列  
file3_columns = ["净资产收益率(平均)(%)", "销售净利率(%)", "销售毛利率(%)", "净利润(元)", "每股收益", "主营营业收入(元)", "总股本", "流通股本"]  
file3 = pd.read_excel("3.xlsx", usecols=file3_columns)  
file3
```

```python
import os  
import pandas as pd  
  
# 设置文件夹路径  
folder_path = 'test1'  
  
# 遍历文件夹中的Excel文件  
for filename in os.listdir(folder_path):  
    if filename.endswith(".xlsx") or filename.endswith(".xls"):  
        file_path = os.path.join(folder_path, filename)  
  
        # 读取Excel文件  
        df = pd.read_excel(file_path)  
  
        # 处理股票代码列  
        df['股票代码'] = df['股票代码'].astype(str).str.zfill(6)  
        df.to_excel(file_path, index=False)  # 如果需要覆盖原文件就使用这一行  
        
```
